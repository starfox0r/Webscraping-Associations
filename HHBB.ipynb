```python
from bs4 import BeautifulSoup
import requests
import time
```

```python
def main(): 
    URL = 'https://www.hamburg.de/branchenbuch/hamburg/10239825/'
    scrape_pages(URL)
    return 0

def scrape_pages(url):
    
    max_pages = 40
    current_page = 0
    
    while current_page <= max_pages:
        print(f'{url}n{current_page}/')

        html = requests.get(f'{url}n{current_page}').text
        soup = BeautifulSoup(html, 'lxml')
        
        container = soup.find_all('li', class_ = 'result-item no-image')
        
        for firm in container:
            company = firm.find('h2', class_ = 'heading__searchresult').text.strip()
            business = firm.find('span').text.strip()
            location = firm.find_all('p', class_ = 'tags-resultlist')[1].text.strip()
            phone = firm.find('li', class_ = 'dropdown-item').text.strip()
            
            try:
                web = firm.find_all('li', class_ = 'dropdown-item')[1].text
            except:
                web = ''
            
            print(f'{company} \n{business} \n{location} \n{phone} \n{web} \n')
        
        time.sleep(10)
        current_page += 20
        print('\n\n')  

if __name__ == '__main__':
    exit(main())
```
