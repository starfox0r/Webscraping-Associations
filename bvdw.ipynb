```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import openpyxl
```

```python
firmlist = []
```

```python
url = 'https://www.bvdw.org/mitgliedschaft/mitgliedsunternehmen/'
page = requests.get(url)
```

```python
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup
```

```python
lists = soup.find_all('ul', class_='memberlist all')
```

```python
lists
```

```python
for content in lists:
    name = content.find_all('li')
    print(name)
```

```python
for raw in name:
    firm = raw.find('li' != 'None')
    firmname = raw.contents[0]
    webpage = raw.contents[1].text
    zf = [firmname, webpage]
    firmlist.append(zf)
    print(zf)
```

```python
print(firmlist)
```

```python
df = pd.DataFrame(firmlist)
```

```python
df.head()
```

```python
df.to_excel('bvdw.xlsx', index=False)
```
