```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
firmlist = []
```

```python
url = 'https://www.vdma.org/mitglieder'
```

```python
data = requests.get(url)
```

```python
soup = BeautifulSoup(data.content, 'html.parser')
```

```python
content = soup.find_all('div', class_ = 'association-member')
```

```python
print(content)
```

```python
for company in content:
    firm = company.find('p', class_ = 'association-member__title').text
    contact = company.find('div', class_ = 'association-member__info-content').text.strip().replace('\n', ' ')
    location = company.find('div', class_ = 'association-member__info-address').text.replace('\n', ' ')
    address = location.replace('Adresse','')
    alle = firm, contact, address
    firmlist.append(alle)
    print(alle)
```

```python
df = pd.DataFrame(firmlist)
```

```python
df.head()
```

```python
df.columns = ['Fima', 'Kontakt', 'Adresse']
```

```python
df.head()
```

```python
df[['eMail', 'Telefon']] = df['Kontakt'].str.split(' ', 1, expand=True)
```

```python
df
```

```python
df['Telefon'] = df['Telefon'].astype('str')
```

```python
df['Telefon']
```

```python
df['Telefon'].str.rsplit(' ', n = 1, expand = True)
```

```python
df[['Phone', 'Webseite']] = df['Telefon'].str.rsplit(' ', n = 1, expand = True)
```

```python
df
```

```python
df = df.drop('Kontakt', axis=1)
df = df.drop('Telefon', axis=1)
```
